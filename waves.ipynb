```python
# Efficient numerical arrays.
import numpy as np
```

```python
# Plotting.
import matplotlib.pyplot as plt
```

```python
# Make plots a little nicer.
plt.style.use('seaborn-deep')
plt.rcParams['figure.figsize'] = (20, 8)
```

```python
# A range of x values for plotting functions.
x = np.linspace(-10.0, 10.0, 1000)
```

```python
# Sin wave.
y_sin = np.sin(x)

# Cos wave.
y_cos = np.cos(x)
```

```python
# Create new plot.
fig, ax = plt.subplots()

# Plot sin wave.
ax.plot(x, y_sin, color='mediumvioletred', label='$sin(x)$')

# Plot cos wave.
ax.plot(x, y_cos, color='rebeccapurple', label='$cos(x)$')

# Draw vertical lines at -pi and pi.
plt.axvline(x=0, color='slategrey', linestyle='--', label='$0$')
plt.axvline(x=2*np.pi, color='seagreen', linestyle='--', label='$2\pi$')

# Show legend.
ax.legend();
```

```python
# Create new plot.
fig, ax = plt.subplots()

# Plot Sin wave.
ax.plot(x, y_sin, color='mediumvioletred', label='$\sin(x)$')

# Plot 2Sin wave.
ax.plot(x, y_2sin, color='rebeccapurple', label='$2 \sin(x)$')

# Draw vertical lines at -pi and pi.
plt.axvline(x=0, color='slategrey', linestyle='--', label='$0$')
plt.axvline(x=2*np.pi, color='seagreen', linestyle='--', label='$2\pi$')

# Show legend.
ax.legend();
```

## Sum of Waves

***

```python
# A range of x values for plotting functions.
x = np.linspace(-np.pi, 6.0 * np.pi, 1000)

# Waves.
ys = np.vstack([
    np.sin(x),
    3.0 * np.sin(x) + 1.0,
    np.sin(x + 1.0)**2,
    np.sin(2.0 * x),
])
```

```python
# Create new plot.
fig, ax = plt.subplots()

for y in ys:
    # Plot component wave.
    ax.plot(x, y, color='lightslategrey', linestyle=':')

# Plot sum wave.
ax.plot(x, ys.sum(axis=0), color='deeppink', label='$\Sigma$')

# Draw vertical lines at -pi and pi.
plt.axvline(x=0, color='teal', linestyle='--')
plt.axvline(x=2*np.pi, color='teal', linestyle='--')

# Show legend.
ax.legend();
```

```python
ax = plt.axes(projection='3d')

x = np.linspace(0, 10.0 * np.pi, 1000)
# f(x) = cos(x) + i sin(x)
y = np.cos(x)
z = np.sin(x)

ax.plot3D(x, y, z);
```

```python
ax = plt.axes(projection='3d')

x = np.linspace(0, 10.0 * np.pi, 1000)
# f(x) = e^ix
y = np.exp(1.0j * x)

ax.plot3D(x, y.real, y.imag);
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 800, cmap='winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
